##Implement Closeness Centrality from scratch (7)

```
#
G = nx.read_edgelist('polblogs.edgelist.simple_format_unweighted', edgetype=int)
```

```
len(G.nodes())
```

```
def closeness_central(graph,node):
  dist_total = 0
  list_nodes = list(graph.nodes())
  n = len(graph.nodes())
  for i in list_nodes:
    try:
      dist_total += nx.shortest_path_length(graph,node,i)
    except:
      pass

  if dist_total <= 1:
    close_central = 0
  else:
    close_central = (n-1) / dist_total
  
  return close_central
```

```
#this is the solution
nx.closeness_centrality(G, 'democraticunderground.com')
```

```
#value is very close could be the disconnected node that's throwing us off
closeness_central(G,'democraticunderground.com')
```

#(2) Implement Page Rank (using Damping factor approach) from scratch (7)

```
list_node = []
for i in list(G.edges()):
  for j in i:
    list_node.append(j)

nodes = set(list_node)
```

```
len(nodes)
```

```
Edges = list(G.edges())
```

```
#create a dictionary of all the edges
graph_dict = {}

for key, val in Edges:
  if key in graph_dict:
    graph_dict[key].append(val)
  else:
    graph_dict[key] = []

for key, val in Edges:
  if val in graph_dict:
    graph_dict[val].append(key)
  else:
    graph_dict[val] = []
```

```
#connected to node

len(graph_dict['tsrightdominion.blogspot.com'])
```

```
#creating a page rank list initializing their PR at 1 first

page_rank = {}
for key in graph_dict:
  page_rank[key] = 1
```

```
page_rank['tsrightdominion.blogspot.com']
```

```
#returns numbr of pgs connected to C(N)
def connect_pg(node):
  return len(graph_dict[node])
```

```
#returns the value of all page rank related to that particular page PR(N)
def page_rank_N(node):
  A = 0
  for i in graph_dict[node]:
    try:
      A += page_rank[i] / connect_pg(i)
    except:
      pass

  return A
```

```
#creating a page rank list initializing their PR at 1 first, required every single time to start pg rank at 1. and keep iterating until the value is not changing much
page_rank = {}
for key in graph_dict:
  page_rank[key] = 1

N = len(graph_dict.keys())

#iterate 100 times
d = 0.85
for i in range(100):
  for key in page_rank:
    page_rank[key] = (1-d)/N + d * page_rank_N(key)

page_rank
```
