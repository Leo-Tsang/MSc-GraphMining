```
def k_prim_maxst(graph_dict,k):

    mst = []
    #start with the largest
    used = set([max(graph_dict, key=graph_dict.get)])
    avail_edges = graph_dict[list(used)[0]]
    sort_edges = sorted(avail_edges, reverse=True)
 
    #here we create our prim mst
    while sort_edges:
    
        siml, a, b = sort_edges[0]
        if b not in used:
            used.add(b)
            mst.append((a,b,siml))
            sort_edges.pop(0)
        
            for i in graph_dict[b]:
                if i not in used:
                    sort_edges.append((i))
                    sort_edges = sorted(sort_edges, reverse=True)
        else:
          sort_edges.pop(0)

    #here we delete our edges based on k value
    k_mst = sorted(mst, key=lambda x: x[2])
    
    for i in range(k-1):
      k_mst.pop(0)

    return k_mst
```

```
#identify edges to be deleted
def del_edges(graph_edge):
  keydel = []
  for key, val in graph_edge.items():
    if val == 'delete':
      keydel.append(key)
  return (keydel)
```

```
import random
def min_cut(graph):

  graph_copy = graph.copy()
  #num_nodes = len(graph_edge)
  used = set()

  #since we know there are 9 nodes altogther, I hard-coded 7 to know it'll stop at 2 remaining
  while len(used) != 7:

    merge_vertex = list(random.choice(list(graph_copy.values())))
    node1 = merge_vertex[0]
    node2 = merge_vertex[1]

    for i in graph_copy:
      old_node = list(graph_copy[i])
      new_node = [x if x != node2 else node1 for x in old_node]
      if new_node[0] == new_node[1]:
        new_node = 'delete'
      graph_copy[i] = new_node

    to_del = del_edges(graph_copy)
    for i in del_edges(graph_copy):
      del graph_copy[i]
  
    used.add(node2)

  cut_edge = []
  for key in graph_copy:
    cut_edge.append(key)

  return(cut_edge)
```
